```python
import pandas as pd
```

```python
def max_depth(node):
    if node is None:
        return 0
    left_depth = max_depth(node.left)
    right_depth = max_depth(node.right)
    
    return 1 + max(left_depth, right_depth)
```

```python
res_dfs = []
```

```python
for r in res:
    df_r = pd.DataFrame([r['summary']])
    r_details = r['fix_details']
    tree_depth_list = [max_depth(r_details[i]['rule'].root) for i in list(r_details)]
    df_r['avg_tree_depth'] = sum(tree_depth_list) / len(tree_depth_list)
    df_r['tree_cnt_after_fix'] = len(r['fix_details'])
    df_r['avg_tree_size'] = sum([x['rule'].size for x in r['fix_details'].values()]) / len(r['fix_details'])
    df_r['runtime_total'] = sum([v for v in r['summary']['runtimes'].values()])
    df_r['repair_time'] = r['summary']['runtimes']['repair_time']
    res_dfs.append(df_r)
```
